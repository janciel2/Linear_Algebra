````python
#1
def matrix_addition(matrix1, matrix2):
    result = []

    for i in range(len(matrix1)):
        row_result = []
        for j in range(len(matrix1[0])):
            row_result.append(matrix1[i][j] + matrix2[i][j])
        result.append(row_result)

    return result
````

````python
#2
def matrix_multiplication(matrix1, matrix2):
    result = []

    for i in range(len(matrix1)):
        row_result = []
        for j in range(len(matrix2[0])):
            element_sum = 0
            for k in range(len(matrix2)):
                element_sum += matrix1[i][k] * matrix2[k][j]
            row_result.append(element_sum)
        result.append(row_result)

    return result
````

````python
#3
def distributive_property(A, B, k):
    # k(A + B)
    kab_result = [[k * (A[i][j] + B[i][j]) for j in range(len(A[0]))] for i in range(len(A))]

    # kA + kB
    ka_result = [[k * A[i][j] for j in range(len(A[0]))] for i in range(len(A))]
    kb_result = [[k * B[i][j] for j in range(len(B[0]))] for i in range(len(B))]
    kab_explicit_result = matrix_addition(ka_result, kb_result)

    return kab_result, kab_explicit_result
````

````python
#4
def extract_diagonal(matrix):
    return [matrix[i][i] for i in range(min(len(matrix), len(matrix[0])))]
````

````python
#5
def matrix_trace(matrix):
    return sum(matrix[i][i] for i in range(min(len(matrix), len(matrix[0]))))
````

````python
# Example usage:
matrix_A = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
matrix_B = [[9, 8, 7], [6, 5, 4], [3, 2, 1]]

# Matrix Addition
result_addition = matrix_addition(matrix_A, matrix_B)
print("Matrix Addition:")
print(result_addition)

# Matrix Multiplication
result_multiplication = matrix_multiplication(matrix_A, matrix_B)
print("\nMatrix Multiplication:")
print(result_multiplication)

# Distributive Property
k = 2
result_distributive, result_distributive_explicit = distributive_property(matrix_A, matrix_B, k)
print(f"\n{k}(A + B) Result:")
print(result_distributive)
print(f"\n{k}A + {k}B Result:")
print(result_distributive_explicit)

# Extract Diagonal
diagonal = extract_diagonal(matrix_A)
print(f"\nDiagonal of Matrix A:")
print(diagonal)

# Matrix Trace
trace_result = matrix_trace(matrix_A)
print(f"\nTrace of Matrix A:")
print(trace_result)
````
